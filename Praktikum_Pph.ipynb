## Data exploration

Explore the loaded data to understand its structure, identify missing values, data types, and the distribution of key variables. Determine the shape of the data and identify potential outliers.


**Reasoning**:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Examine data types
print("Data Types:\n", df.dtypes)

# 2. Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values)
print("\nColumns with Missing Values:\n", df.columns[df.isnull().any()])

# Visualize missing values using a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# 3. Analyze numerical variables
numerical_cols = df.select_dtypes(include=['number']).columns
print("\nDescriptive Statistics for Numerical Variables:\n", df[numerical_cols].describe())

for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    df[col].plot(kind='hist', title=f'Distribution of {col}')
    plt.show()
    plt.figure(figsize=(8, 4))
    df.boxplot(column=col)
    plt.title(f'Boxplot of {col}')
    plt.show()

# Identify potential outliers (example using IQR)
potential_outliers = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    potential_outliers[col] = outliers[col].tolist()

print("\nPotential Outliers:\n", potential_outliers)


# 4. Explore categorical variables
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\nUnique values and frequencies for {col}:\n{df[col].value_counts()}")
    plt.figure(figsize=(8, 4))
    df[col].value_counts().plot(kind='bar', title=f'Distribution of {col}')
    plt.show()

# 5. Summarize findings
print("\nData Shape:", df.shape)
# ... (Summarize other findings)
```

## Data cleaning

### Subtask:

Handle missing values, address inconsistencies, remove duplicates, handle outliers, and convert data types as instructed in the subtask.



```python
# Handle Missing Values
for col in df.select_dtypes(include=['number']).columns:
    df[col] = df[col].fillna(df[col].median())
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna(df[col].mode()[0])

# Address inconsistencies in categorical columns
# Example: Standardize 'YES' and 'yes' to 'yes'
for col in ['kebijakan_tax_holiday_awal', 'kebijakan_tax_holiday_akhir', 'aset_kategori', 'aset_metode', 'transaksi_skenario']:
    if col in df.columns:  # Check if the column exists
        df[col] = df[col].str.lower()

# Convert 'aset_nilai_perolehan' to numeric, handle errors, and impute
if 'aset_nilai_perolehan' in df.columns:
    df['aset_nilai_perolehan'] = pd.to_numeric(df['aset_nilai_perolehan'], errors='coerce')
    df['aset_nilai_perolehan'] = df['aset_nilai_perolehan'].fillna(df['aset_nilai_perolehan'].median())

# Remove duplicates
df.drop_duplicates(inplace=True)

# Outlier handling (example using IQR, adjust as needed)
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)  # Winsorizing outliers

# Data type conversion (if needed)
# ... (Add code to convert other columns to appropriate data types)
display(df.head())
display(df.info())
```

## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Create new features, transform data types, and verify the changes.



```python
# Feature Engineering: Calculate asset age and profitability
df['aset_umur'] = 2025 - df['aset_tahun_perolehan']  # Calculate asset age as of 2025
df['transaksi_profitabilitas'] = df['transaksi_pendapatan'] - df['transaksi_beban_operasional']

# Data Transformation: Convert relevant columns to appropriate data types
# Convert to numeric, coercing errors to NaN
for col in ['aset_nilai_perolehan', 'aset_umur_ekonomis', 'aset_tahun_perolehan', 'transaksi_tahun', 'transaksi_pendapatan', 'transaksi_beban_operasional', 'transaksi_penyusutan']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Impute missing values after type conversion
for col in ['aset_nilai_perolehan', 'aset_umur_ekonomis', 'aset_tahun_perolehan', 'transaksi_tahun', 'transaksi_pendapatan', 'transaksi_beban_operasional', 'transaksi_penyusutan']:
    df[col] = df[col].fillna(df[col].median())

# One-hot encode categorical features
categorical_cols = ['aset_kategori', 'aset_metode', 'transaksi_skenario']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Verification
display(df.head())
display(df.info())
```
